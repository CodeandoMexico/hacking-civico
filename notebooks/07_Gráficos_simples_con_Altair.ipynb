# Gráficos simples con Altair

```
import altair as alt
import seaborn as sns
```

```
datos = sns.utils.load_dataset("penguins")
```

```
datos.head()
```

```
alt.Chart(datos).mark_line().encode(
    x = "bill_depth_mm",
    y = "body_mass_g",
    color = "species"
)
```

```
alt.Chart(datos).mark_circle(size = 100, fillOpacity = .50).encode(
    x = alt.X("bill_depth_mm:Q", title = "Profundidad del pico (mm)"),
    y = alt.Y("body_mass_g:Q", title = "Masa corporal (g)"),
    color = alt.Color("species:N", title = "Especies"),
).properties(
    title = "Masa corporal vs profundidad de pico",
    height = 200,
    width = 200,
)
```

```
from vega_datasets import data

empleos = data.us_employment()
```

```
empleos.head()
```

```
alt.Chart(empleos).mark_bar().encode(
    x = alt.X("month:T"),
    y = alt.Y("nonfarm_change:Q")
).properties(
    title = "Cambio en el sector no-agrícola de Estados Unidos",
    width = 600
)
```

El desafío es crear un gráfico que:
1. Utilice datos de empleo de estados unidos (`data.us_employment()`).
2. Sea un gráfico de línea.
3. En el eje $x$ visualices los meses.
4. En el eje $y$ visualices el cambio en el sector no agricola (`nonfram_change`).
5. Que la línea sea de color negro.
6. Que el gráfico sea de un ancho de 650px.
7. Que el gráfico tenga un título descriptivo.
8. Bonus: que el eje $x$ no tenga título.

Bonus:
El mismo gráfico pero en "área", es decir, que sea un gráfico de área.
